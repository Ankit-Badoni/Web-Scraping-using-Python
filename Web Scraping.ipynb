```python
import requests as rt
import bs4
```

```python
key=input("Enter The Problem Or Medicine Name = ")
url="https://www.1mg.com/search/all?name="+key
url
```

```python
# user agent act as a web browser to get content from the website if website rejects the request  
# of user to fetch HTML content.

h={'User-agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.106 Safari/537.36 OPR/38.0.2220.41'}
```

```python
dt=rt.request("get",url,headers=h)
```

```python
dt.status_code
```

```python
# showing all the HTML text

dt.text
```

```python
dt.text
```

```python
sp=bs4.BeautifulSoup(dt.text,"html.parser")
```

```python
sp.text
```

```python
num=0
List=[]
for i in sp.findAll("div",{'class':'col-md-3 col-sm-4 col-xs-6 style__container___jkjS2'}):
    n=i.find("div",{'class':'style__pro-title___3G3rr'}).text
    p=i.find("div",{'class':'style__price-tag___KzOkY'}).text
    p=p.replace("MRP",'')
    p=p.lstrip()
    link=i.find('a').get("href")
    l="https://www.1mg.com"+link
    num=num+1
    print("------------------------------------Number-",num,"----------------------------------------")
    print("NAME= ",n,end='\n \n')
    print("PRICE= ",p,end='\n \n')
    print("LINK=",l,end='\n \n')
    dt2=rt.request("get",l,headers=h)
    sp2=bs4.BeautifulSoup(dt2.text,"html.parser")
    
    try:
        info=sp2.find("div",{'class':'ProductDescription__description-content___A_qCZ'}).text
        print('INFORMATION=',"\n \n",info,end='\n \n \n \n')
    except:
        print("Description not available")
    List.append([n,p,l,info])
```

```python
import pandas as pd
```

```python
data=pd.DataFrame(List,columns=[['Name','Price','Link','Description']])
```

```python
data.head()
```

```python
data.to_csv('data.csv')
```
